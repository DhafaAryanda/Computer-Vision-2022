```python
#Berikut ini adalah code untuk menampilkan video live stream dari kamera 
#framenya diperkecil menjadi setengahnya
#untuk menutup frame, klik key 'q'

import cv2

cap = cv2.VideoCapture(0)

# Check if the webcam is opened correctly
if not cap.isOpened():
    raise IOError("Cannot open webcam")
    
while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
    cv2.imshow('Input', frame)
    
    if cv2.waitKey(1) & 0xff == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
# Melakukan blur pada wajah dengan level blur yang berbeda beda dengan Gaussian

import cv2
import numpy as np

def print_howto(): 
    print("""
        Change mode of image:
            1. Blur min - press '1'
            2. Blur med - press '2'
            3. Blur max - press '3'
            """)

# download file xml: https://raw.githubusercontent.com/kassumanette/viskom/main/haarcascade_frontalface_default.xml
face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

print_howto()
cap = cv2.VideoCapture(0)

blur1 = False
blur2 = False
blur3 = False

while True:
    ret, frame = cap.read()
    if(ret):
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        f = face.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 7)
        
        for x, y, w, h in f:
            if blur1:
                sub_face = frame[y:y+h, x:x+w]
                sub_face = cv2.GaussianBlur(sub_face, (15,15), 0)
                frame[y:y+sub_face.shape[0], x:x+sub_face.shape[1]] = sub_face
                
            if blur2:
                sub_face = frame[y:y+h, x:x+w]
                sub_face = cv2.GaussianBlur(sub_face, (25,25), 10)
                frame[y:y+sub_face.shape[0], x:x+sub_face.shape[1]] = sub_face
                
            if blur3:
                sub_face = frame[y:y+h, x:x+w]
                sub_face = cv2.GaussianBlur(sub_face, (45,45), 40)
                frame[y:y+sub_face.shape[0], x:x+sub_face.shape[1]] = sub_face                
            
        cv2.rectangle(frame, (x,y), (x+h,y+w), (0,0,255),2)  # frame
        cv2.imshow('Webcam', frame) #membuka webcam
    ch = 0xFF &cv2.waitKey(1)
    
    if ch == ord("1"):
        blur1 = not blur1
        
    if ch == ord("2"):
        blur2 = not blur2
        
    if ch == ord("3"):
        blur3 = not blur3
        
    if ch == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
